```python
import os
import numpy as np
import torch
from collections import defaultdict
from Sims import get_loaders, load_Model

def get_data_distribution(configs):
    # 使用现有的get_loaders函数加载数据集
    ClientLoaders, client_testloaders, TrainLoader, TestLoader, stats = get_loaders(
        configs['dname'], configs['nclients'], configs['isIID'], configs['alpha'], 
        configs['aug'], False, False, configs['normal'], configs['data_shuffle'], 
        configs['batch_size']
    )
    
    # 初始化一个字典来存储每个客户端的数据分布
    client_data_distribution = defaultdict(lambda: defaultdict(int))

    # 遍历每个客户端的数据加载器
    for client_id, loader in enumerate(ClientLoaders):
        for inputs, targets in loader:
            # 统计每个类别在目标中的出现次数
            for target in targets:
                client_data_distribution[client_id][target.item()] += 1

    return client_data_distribution

if __name__ == '__main__':
    # 示例配置
    Configs = {
        'dname': 'cifar10',
        'nclients': 128,
        'isIID': False,
        'alpha': 0.1,
        'aug': False,
        'normal': True,
        'data_shuffle': True,
        'batch_size': 16,
    }
    
    # 获取数据分布
    distribution = get_data_distribution(Configs)
    distribution_dict = {}
    # 打印每个客户端的数据分布
    for client_id, dist in distribution.items():
        distribution_dict[client_id] = dict(dist)
        # print(f"客户端 {client_id}: {dict(dist)}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# 假设这里的 distribution_dict 是您之前提到的数据分布字典
# 客户端数据分布字典
data_distribution = distribution_dict

# 将数据转化为pandas DataFrame
df = pd.DataFrame(data_distribution).fillna(0).astype(int).T
df.columns = [f'Class {i}' for i in range(10)]  # 为类别命名

# 设置绘图风格
sns.set(style="whitegrid", palette="muted")

# 创建堆叠条形图
fig, ax = plt.subplots(figsize=(14, 8))
bottom = np.zeros(len(df))
colors = sns.color_palette("Set3", n_colors=10)

for i, col in enumerate(df.columns):
    ax.bar(df.index, df[col], bottom=bottom, color=colors[i], label=col)
    bottom += df[col]

# 设置图例、标题和标签
ax.set_title('Client Data Distribution - CIFAR-10', fontsize=20)
ax.set_xlabel('Client ID', fontsize=15)
ax.set_ylabel('Number of Samples', fontsize=15)
ax.legend(title='Classes', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.xticks(rotation=90)

# 调整布局和显示图像
plt.tight_layout()

# 保存图像
plt.savefig('client_data_distribution.png')

plt.show()
```
