```python
import re
import numpy as np
import matplotlib.pyplot as plt

def extract_ptest_acc_from_log(file_path):
    ptest_acc_list = []
    
    with open(file_path, 'r') as file:
        lines = file.readlines()
    
    for line in lines:
        match = re.search(r'ptest_acc:([\d.]+)', line)
        if match:
            ptest_acc_list.append(float(match.group(1)))
    
    return ptest_acc_list
def calculate_stats(ptest_acc_list):
    if len(ptest_acc_list) < 10:
        return None, None
    last_10 = ptest_acc_list[-10:]
    mean = np.mean(last_10)
    std_dev = np.std(last_10)
    return mean, std_dev
def plot_ptest_acc(ptest_acc_list):
    plt.figure(figsize=(10, 5))
    plt.plot(ptest_acc_list, marker='o', linestyle='-', color='b')
    plt.title('ptest_acc Over Time')
    plt.xlabel('Train Round')
    plt.ylabel('ptest_acc')
    plt.grid(True)
    plt.show()
```

```python
log_file_path = './fedcea_tab3/20240531_181659/fl_log.txt'
ptest_acc_list = extract_ptest_acc_from_log(log_file_path)
mean, std_dev = calculate_stats(ptest_acc_list)
print(f'Mean of last 10 ptest_acc: {round(100*mean,2)}')
print(f'Standard Deviation of last 10 ptest_acc: {round(100*std_dev,2)}')

# Plot the ptest_acc list
plot_ptest_acc(ptest_acc_list)
```

```python
import re
from datetime import datetime

def extract_times_and_calculate_difference(index,file_path):
    # 读取log文件内容
    with open(file_path, 'r') as file:
        log_data = file.readlines()
    
    # 提取“Time”索引的值
    times = []
    for line in log_data:
        if 'Time:' in line:
            # 正则表达式匹配Time索引的时间戳
            time_match = re.search(r'Time: (\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2})', line)
            if time_match:
                times.append(time_match.group(1))
    
    # 确保有足够的时间戳
    if len(times) < index:
        raise ValueError(f"Log data does not contain at least {index} 'Time' timestamps.")
    
    # 转换为datetime对象
    times = [datetime.strptime(time, '%Y-%m-%d %H:%M:%S') for time in times]
    
    # 计算第25个值和第一个值的差
    time_difference = times[index-1] - times[0]
    
        # 将时间差转换为分钟
    time_difference_in_minutes = time_difference.total_seconds() / 60
    
    return times, time_difference_in_minutes

# 替换为你的log文件路径
# log_file_path = './fedcea_tab3/20240531_182736/fl_log.txt'
# log_file_path = './fedcea_tab3/20240531_182001/fl_log.txt'
log_file_path = './fedcea_tab3/20240531_181659/fl_log.txt'

times, time_difference = extract_times_and_calculate_difference(100,log_file_path)
# print("Times:", times)
print("Difference between the 25th and the 1st time:", time_difference)

def calculate_differences(participants):
    differences = []
    for i in range(1, participants.size()):
        time_differe = ticipantstime_difference - participants[i - 1]
        time_difference_in_minutes = time_difference.total_seconds() / 60
        print(time_difference_in_minutes)
        differences.append(time_difference_in_minutes)
        
    return differences


# 计算前后两个元素的差值并放到新的列表中
differences_list = calculate_differences(participants)
print("Differences:", differences_list)

```

```python
#读取参数数量】 每一轮 16 up down 
import re

def extract_participants(file_path):
    participants = []
    with open(file_path, 'r') as file:
        log_data = file.readlines()
        
    for line in log_data:
        if 'Participants:' in line:
            # 正则表达式匹配Participants后的值
            participants_match = re.search(r'Participants: (\d+)', line)
            if participants_match:
                participants.append(int(participants_match.group(1)))
    
    return participants

def mean_first_25_participants(participants):
    # 确保有足够的值
    if len(participants) < 25:
        raise ValueError("The list does not contain at least 25 values.")
    
    # 计算前25个值的均值
    mean_value = sum(participants[:25]) / 25
    return mean_value

# 替换为你的log文件路径
log_file_path = './fedcea_tab3/20240531_182736/fl_log.txt'
participants_list = extract_participants(log_file_path)
print("Participants:", participants_list)

# 计算前25个值的均值
mean_of_first_25 = mean_first_25_participants(participants_list)
print("Mean of the first 25 participants:", mean_of_first_25)
```
